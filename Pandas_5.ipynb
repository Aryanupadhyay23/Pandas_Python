```python
import numpy as np
import pandas as pd
```

```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
# The problem?
a[('cse',2022)]
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels[0]
# 2. pd.MultiIndex.from_product()
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

```python
# level inside multiindex object
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
print(s)
```

```python
s[('cse',2022)]
```

```python
s['cse']
```

```python
temp = s.unstack() # unstack converts the multi-index series into a dataframe
temp
```

```python
temp.stack() # it converts the dataframe into multi-index series back
```

### Long Vs Wide Data
image.png

**Wide format** is where we have a single row for every data point with multiple columns to hold the values of various attributes.

**Long format** is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.

```python
pd.DataFrame({'cse':[120]})
```

```python
# melt is used to convert from wide to long
pd.DataFrame({'cse':[120]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]})
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(value_name='num_students', var_name='branch')
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
)
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
).melt()
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
).melt(id_vars=['branch'],var_name='year',value_name='students')
```

### Pivot Table

The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
df.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='count')
```

```python
df.pivot_table(index='sex',columns='smoker',aggfunc='count')
```

```python
df.pivot_table(index='sex',columns='smoker',aggfunc='count')['tip']
```

```python
df.pivot_table(index='sex',columns='smoker',aggfunc='count')['size']
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill')
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'})
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'},margins=True)
```

```python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum',margins=True)
```
